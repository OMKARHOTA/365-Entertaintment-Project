````python
#here we are try to find whether the higher Annual income is proportional to the higher credit score or not
filtered_merged_row = df[(df['Annual_Income'] > 100000) | (df['Credit_Score'] == 'Good')]
print(filtered_merged_row)
````

````python
#Does the credit utilization ratio is impact on credit scores as compared to total number of loans taken
Credit_Utilization_Ratio = df['Credit_Utilization_Ratio'].reset_index(drop=True) # Reset index for alignment
Credit_Score = df['Credit_Score'].reset_index(drop=True) # Reset index for alignment

merged_row = pd.concat([Credit_Utilization_Ratio, Credit_Score], axis=1) # Concatenate horizontally
print(merged_row)
````

````python
#now comparing credit score with number of loan
Num_of_Loan = df['Num_of_Loan'].reset_index(drop=True) # Reset index for alignment
Credit_Score = df['Credit_Score'].reset_index(drop=True) # Reset index for alignment

merged_row = pd.concat([Num_of_Loan, Credit_Score], axis=1) # Concatenate horizontally
print(merged_row)
````

````python
#is having higher number of credits card associated with credit score assumin the credit managing ratio is well?
Num_of_Loan = df['Num_Credit_Card'].reset_index(drop=True) # Reset index for alignment
Credit_Score = df['Credit_Score'].reset_index(drop=True) # Reset index for alignment

merged_row = pd.concat([Num_of_Loan, Credit_Score], axis=1) # Concatenate horizontally
print(merged_row)
````

````python
#is there any negative corelation between the number of delayed payments and credit scores
Num_of_Delayed_Payment = df['Num_of_Delayed_Payment'].reset_index(drop=True) # Reset index for alignment
Credit_Score = df['Credit_Score'].reset_index(drop=True) # Reset index for alignment

merged_row_4 = pd.concat([Num_of_Delayed_Payment, Credit_Score], axis=1) # Concatenate horizontally
print(merged_row_4)
````
